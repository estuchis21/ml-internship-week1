## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

**What this work CAN say:**
- We observed that pages with certain patterns (stale updates, declining
  trend, visible traffic) tend to score higher in a review queue.
- These are directional, decision-support signals — useful for helping a
  reviewer prioritize limited time, not a certainty about any single page.
- Any relationship reported is observational: we watched what happened,
  we did not run an experiment.

**What this work will NEVER say:**
- That refreshing a page *caused* a traffic recovery — that requires a
  controlled experiment (A/B test) or another causal design, not historical
  data alone.
- That we've discovered or reproduced a Google ranking algorithm factor.
- That a "declining" label is a guaranteed real decline — it could instead be
  seasonality, consolidation to a sibling page, or plain noise.
- That the model's score is "the truth" — it's a ranked hypothesis for a
  human to check, not a final verdict.
